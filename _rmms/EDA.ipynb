```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

port2Features = port2[['temp','press','tot_ton','tot_cal','freq','net_cal','conduc','speed']]
port2['stat'] = port2.net_kg < 10000
port2Labels = port2['net_kg']#port2['stat']

model = DecisionTreeClassifier(
    criterion='entropy',  #'entropy','gini'
    max_depth=5
    ).fit(X=port2Features
         ,y=port2Labels)

print(model)
```

```python
import graphviz 

dot_data = tree.export_graphviz(
    model
,   out_file=None
,   feature_names=['temp','press','tot_ton','tot_cal','freq','net_cal','conduc','speed']
,   class_names=port2['net_kg'] #['bad','good']
)
graph = graphviz.Source(dot_data)
graph.render("decition-tree") 
graph
```
