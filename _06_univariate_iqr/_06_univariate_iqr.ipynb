````python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())

df = pd.read_csv("../data_source/WIP_HISTORY.csv")

````

````python
df.head()
````

````python
df1 = pysqldf("select LINE, GAS_PRESSURE from df ;")
box = df1.boxplot(column=['GAS_PRESSURE'], by='LINE')
````

````python
df1 = pysqldf("select LINE, THICKNESS from df ;")
box = df1.boxplot(column=['THICKNESS'], by='LINE')
````

````python
df1 = pysqldf("select LINE, TIME_GAP, RPM from df ;")
box = df1.boxplot(column=['TIME_GAP','RPM'], by='LINE')
````

````python
box = df1.boxplot(column=['TIME_GAP'], by='LINE')
````

````python
df1 = pysqldf("select LINE, TIME_GAP, RPM from df ;")
box = df1.boxplot(column=['RPM' , 'TIME_GAP'], by='LINE')
````
