```python
# -*- coding: utf-8 -*-
"""
[Title] Histogram
[Author] Yibeck Lee(yibec.Lee@gmail.com)
[Contents] 
 - Histogram & Descriptive Statistics(기술통계)
"""
# %matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

title = '[기술통계량(TEMPERATURE)]'
param = 'TEMPERATURE' 

df = pd.read_csv("WIP_HISTORY.csv")
print('[기술통계량(TEMPERATURE)]')
df['TEMPERATURE'].describe()
```

```python
#Summary File Saving
data = df.describe()
data.to_csv("summary.csv")
```

```python

plt.rcParams["figure.figsize"] = (20, 10)

plt.subplot(1,4,1)
plt.title("Histogram(Gas Pressure. bins=3)")
plt.ylabel("frequency")
plt.xlabel("gas Pressure")
df['GAS_PRESSURE'].hist(bins=3)

plt.subplot(1,4,2)
plt.title("Histogram(Gas Pressure. bins=5)")
plt.ylabel("frequency")
plt.xlabel("gas Pressure")
df['GAS_PRESSURE'].hist(bins=5)

plt.subplot(1,4,3)
plt.title("Histogram(Gas Pressure. bins=10)")
plt.ylabel("frequency")
plt.xlabel("gas Pressure")
df['GAS_PRESSURE'].hist(bins=10)

plt.subplot(1,4,4)
plt.title("Histogram(Gas Pressure. bins=20)")
plt.ylabel("frequency")
plt.xlabel("gas Pressure")
df['GAS_PRESSURE'].hist(bins=20)

plt.show()
```
