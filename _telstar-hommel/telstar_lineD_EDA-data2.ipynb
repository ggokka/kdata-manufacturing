```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_excel('telstar_line_d_data2.xlsx', sheet_name='part')
df.head()
```

```python
df.describe()
```

```python
df_notnull = df.dropna();
```

```python
df_notnull.describe()
```

```python
# 2-1. Multiple histograms on the same axis

var_color_dict = {'S/RING_Assy_DATA': 'blue', 
                      'SEAL_PRESS_LOAD': 'red', 
                      'SEAL_PRESS_DISTANCE': 'yellow', 
                      'N/R_DATA#1': 'green'}
# for loop
for var in var_color_dict:
    sns.distplot(df_notnull[var],
                    color = var_color_dict[var],
                    hist_kws = {'edgecolor': 'gray'},
                    label = var)

plt.legend(title = 'Variables')
plt.show()

#출처: https://rfriend.tistory.com/409 [R, Python 분석과 프로그래밍 (by R Friend)]
```

```python
bins = np.linspace(-10, 20, 100)

plt.hist(df_notnull['S/RING_Assy_DATA'], bins, alpha=0.5, label='S/RING_Assy_DATA')
plt.hist(df_notnull['SEAL_PRESS_LOAD'], bins, alpha=0.5, label='SEAL_PRESS_LOAD')
plt.legend(loc='upper right')
plt.show()
```

```python
for v in df_notnull.columns:
    plt.hist(df_notnull[v], label=v)
#plt.hist(df_notnull['SEAL_PRESS_LOAD'], label='SEAL_PRESS_LOAD')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.hist(df_notnull['SEAL_PRESS_LOAD'], label='SEAL_PRESS_LOAD')
plt.hist(df_notnull['SEAL_PRESS_DISTANCE'], label='SEAL_PRESS_DISTANCE')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.hist(df_notnull['N/R_DATA#1'], label='N/R_DATA#1')
plt.hist(df_notnull['N/R_DATA#2'], label='N/R_DATA#2')
plt.hist(df_notnull['N/R_DATA#3'], label='N/R_DATA#3')
plt.legend(loc='upper right')
plt.show()
```
